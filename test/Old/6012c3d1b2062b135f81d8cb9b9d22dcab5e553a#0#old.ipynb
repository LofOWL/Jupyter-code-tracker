```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline

# plt.style.use('gadfly')
# plt.rcParams['font.size'] = 18
```

```python
data = pd.DataFrame({'A Ratings': [6.5, 8.6], 'B Ratings': [8.1, 9.1], 
                     'A Number': [10, 300], 'B Number': [180, 40], 
                     'Gender': ['M', 'F']}).set_index('Gender')
data
```

```python
a_average = (data.loc[:, 'A Ratings'] * data.loc[:, 'A Number']).sum() / data.loc[:, 'A Number'].sum(axis=0)
b_average = (data.loc[:, 'B Ratings'] * data.loc[:, 'B Number']).sum() / data.loc[:, 'B Number'].sum(axis=0)


data.loc['Combined'] = [a_average, 
                        b_average, 
                        data['A Number'].sum(), 
                        data['B Number'].sum()]
data.round(2)
```

# Hours of Exercise vs Chance of a Heart Attack

```python
n_samples = 100
ages = np.random.randint(20, 50, n_samples)
hours = np.random.randint(2, 10, n_samples) + np.random.randn(n_samples)
p = 15 + 0.5 * ages + -0.7 * hours + np.random.randn(n_samples) *  10
under_50 = pd.DataFrame({'age': ages, 'hours': hours, 'probability': p})
```

```python
under_50.plot(x = 'age', y = 'probability', kind = 'scatter', title = 'P vs Age');
```

```python
under_50.plot(x = 'hours', y = 'probability', kind = 'scatter')
```

```python
plt.plot(under_50['age'], under_50['probability'], marker = 'o')
plt.title('P vs Age');
```
